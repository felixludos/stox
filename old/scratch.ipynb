```python
def get_startdate(yrs):
    return datetime.today() - timedelta(days=int(365*yrs))
def get_ticker(ticker):
    return yf.Ticker(ticker)
```

```python
def get_closes(ticker, yrs=None, p=True):
    start = get_startdate(yrs) if yrs is not None else None
    data = get_ticker(ticker).history(start=start).to_numpy()[:,3]
    if p:
        data /= data[0]
        data -= 1
        data *= 100
    return data
```
